```python
import os
import time
from selenium import webdriver



    


def iter_png (path) :
    names = []
    for fn in os.listdir(path):
         if  os.path.isfile(path+'/'+fn)  :
            if fn.split('.')[-1]=='html' :
                names+=[fn]
         else :
            if path =='.' :
                iter_png (fn)
            else :
                iter_png (path+'/'+fn)
    s = path+'\n\n'
    s += ('<select onchange="theThingToDoIfItChange()" id="selection">\n')
    for n in names :
        n = '.'.join(n.split('.')[0:-1])
        if ('_' in n) :
            n = n.split('_')[1]
        s += '      <option value="'+n+'">'+n+'</option>\n'
    s += '</select>\n\n'
    
    s+='var dict = {\n'
    for n in names :
        n = '.'.join(n.split('.')[0:-1])
        if ('_' in n) :
            n = n.split('_')[1]
        for dec in range(1980,2020,10) :
            
            s += '      "'+n+'/'+str(dec)+'": "describe '+n+' for ' + str(dec)+'"'+',\n'
    s += '      };\n\n\n\n_______________\n'
    print (s)
         
iter_png('../docs/assets/data')
```
